# Project Crawl Data từ web

## Import các thư viện cần thiết

```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
```

## Khai báo đường dẫn

```python
url = 'https://finance.yahoo.com/markets/stocks/most-active/'
```

## Tiến hành lấy response từ đường dẫn

```python
res = requests.get(url)
```

```python
res
```

- Sau khi nhận được response status là 200 điều này có nghĩa là yêu cầu đưa đên server của trang web được chấp nhận

## Lấy phần text từ phản hồi nhận được từ server thông qua Thư viện bs4 và hàm BeautifulSoup

```python
soup = BeautifulSoup(res.text)
```

## Lọc Data có tag là tbody và tr từ text nguồn

```python
data_html = soup.find('tbody').findAll('tr')
```

## In ra để xem dạng của data_html

```python
data_html
```

## TIến hành tạo vòng lặp để lọc các dòng trong tag td từ data_html

```python
data_raw =[]
for i in range(0, len(data_html)):
    data_raw.append(data_html[i].find('td'))
    
```

## In ra data_raw để kiểm tra

```python
data_raw
```

## Lọc ra các dòng trong tag a từ data_raw

```python
data = []
for i in data_raw:
    data.append(i.find('a'))
```

## Tạo 1 vòng lặp để thêm các giá trị cần lấy vào các list rỗng đã cho từ trước từ data_html

```python
Name = []
Price = []
Volume = []
Avg_Volume = []
Market_Cap = []
PE_Ratio = []
WeekChange = []
for i in data_html:
    Name.append(i.find('td').text)
    Price.append(i.findAll('td')[3].text)
    Volume.append(i.findAll('td')[6].text)
    Avg_Volume.append(i.findAll('td')[7].text)
    Market_Cap.append(i.findAll('td')[8].text)
    PE_Ratio.append(i.findAll('td')[9].text)
    WeekChange.append(i.findAll('td')[10].text)
```

## Chuyển các list đó về dạng bảng và đặt tên cho từng cột

```python
data = pd.DataFrame({'Name': Name, 'Price': Price, 'Volume': Volume, 'Avg_Volume': Avg_Volume, 'Market_Cap': Market_Cap, 'PE_Ratio': PE_Ratio, 'WeekChange': WeekChange}, index = None)
```

## In ra data để kiểm tra

```python
data
```

## Tiến hành clean data đã có

### Thay thế các phần không có giá trị thành khoảng trắng

```python
data['Name'] = data['Name'].str.replace('\n', '')
```

```python
data['Price'] = data['Price'].str.replace('\n', '')
```

### Chia cột

#### Chia cột name theo khoảng trắng và lọc ra giá trị Name cần dùng

```python
data['Name']=data['Name'].str.split('      ', expand = True, )[0].reset_index(drop=True)
```

#### In ra data để kiểm tra

```python
data 
```

#### Chia tiếp cột Price thành các cột Price_New, Change, Change_Percent

```python
data['Price_New'] = data['Price'].str.split(' ', expand = True, )[1].reset_index(drop=True)
data['Change'] = data['Price'].str.split(' ', expand = True, )[2].reset_index(drop=True)
data['Change_Percent'] = data['Price'].str.split(' ', expand = True, )[3].reset_index(drop=True)
```

#### In ra data để kiểm tra

```python
data 
```

### Xoá cột

#### Xoá cột Price ban đầu

```python
data = data.drop(columns = ['Price'])
```

#### In lại data để kiểm tra

```python
data 
```

## Lưu lại thành dạng csv để tiếp tục sử dụng

```python
data.to_csv('Most_Active_Stocks.csv', index = False)
```
